## Generar reportes PDF con texto, imágenes y tablas

Documentación: https://py-pdf.github.io/fpdf2/Tutorial-es.html

### Importar la librería `fpdf2`

```python
# pip install fpdf2
from fpdf import FPDF
```

### Crear un PDF

```python
pdf = FPDF()
```

### Agregamos una nueva hoja

```python
pdf.add_page()
```

### Ajustamos la fuente

```python
pdf.set_font("helvetica", size=12)
```

### Agregar una imagen

```python
pdf.image("cliente_thumb.png", x=10, y=8, w=40)
```

### Agregar un salto de línea

```python
pdf.ln(40)
```

### Agregar los datos de una tabla

```python
tabla = [
    ["Nombre", "Edad", "Sexo"],
    ["Ana", "23", "MUJER"],
    ["Beto", "32", "HOMBRE"],
    ["Carla", "28", "MUJER"],
]

for i, fila in enumerate(tabla):
    for j, columna in enumerate(fila):
        if i == 0:
            style = "B"
        elif j == 0:
            style = "BI"
        elif j == 1:
            style = "I"
        else:
            style = ""
        pdf.set_font(style=style)
        pdf.cell(pdf.w / 4, 10, text=columna, border=1)
    pdf.ln(10)
```

```python
pdf.output("Reporte.pdf")
```
