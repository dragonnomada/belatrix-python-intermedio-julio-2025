```python
edades = [23, 45, 62, 47, 29, 17, 24, 21]

edades
```

```python
edad_objetivo = 62

n = len(edades)

for i in range(n):
  edad = edades[i]
  if edad_objetivo == edad:
    print(f"{edad} <-- Encontrada")
  else:
    print(edad)
```

```python
def contiene(serie, x):
  for s in serie:
    if s == x:
      return True
  return False
```

```python
contiene(edades, 55)
```

```python
contiene(edades, 17)
```

```python
def indice(serie, x):
  for i, s in enumerate(serie):
    if s == x:
      return i
  return None
```

```python
indice(edades, 55)
```

```python
indice(edades, 17)
```

```python
edades, edades[5]
```

```python
def buscar(serie, x):
  for i, s in enumerate(serie):
    if s == x:
      return i, s
  return None
```

```python
buscar(edades, 55)
```

```python
buscar(edades, 17)
```

```python
def buscarTodos(serie, x):
  encontrados = []
  for i, s in enumerate(serie):
    if s == x:
      encontrados.append(i)
  return encontrados
```

```python
buscarTodos(edades, 55)
```

```python
edades
```

```python
buscarTodos(edades, 17)
```

```python
productos = [
    {
        "nombre": "Coca Cola",
        "precio": 18.50,
        "categorias": ["refresco", "bebida"]
    },
    {
        "nombre": "Gansito",
        "precio": 21.50,
        "categorias": ["pan", "dulce", "chocolate"]
    },
]
```

```python
def buscarProductoNombre(productos, nombre):
  for producto in productos:
    if producto["nombre"] == nombre:
      return producto
```

```python
buscarProductoNombre(productos, "Coca Cola")
```

```python
def buscarProductoPrecioInferior(productos, precioInferior):
  for producto in productos:
    if producto["precio"] >= precioInferior:
      return producto

def buscarProductoPrecioSuperior(productos, precioSuperior):
  for producto in productos:
    if producto["precio"] <= precioSuperior:
      return producto
```

```python
buscarProductoPrecioInferior(productos, 20)
```

```python
buscarProductoPrecioSuperior(productos, 20)
```

```python
def buscarPorCategoria(productos, categoria):
  for producto in productos:
    if contiene(producto["categorias"], categoria):
      return producto
```

```python
buscarPorCategoria(productos, "bebida")
```

```python
buscarPorCategoria(productos, "chocolate")
```

## Análisis en el rendimiento de los estudiantes

```python
import pandas

students = pandas.read_csv("https://github.com/dragonnomada/belatrix-python-intermedio-julio-2025/raw/refs/heads/main/datos/student_habits_performance.csv")

students.head()
```

```python
import seaborn

seaborn.jointplot(x=students["study_hours_per_day"], y=students["exam_score"])
```

```python
students["age"]
```

```python
# [ T(x) for x in seq ]

# [edad ** 2 for edad in students["age"]]

def grupo_edad(edad):
  if edad <= 18:
    return "A"
  if edad <= 20:
    return "B"
  if edad <= 22:
    return "C"
  return "D"

students["age_group"] = students["age"].map(grupo_edad)

students
```

```python
reporte1 = students[["age_group", "exam_score"]]

seaborn.barplot(reporte1, x="age_group", y="exam_score", hue=students["gender"])
```

```python
seaborn.barplot(reporte1, x="age_group", y="exam_score", hue=students["parental_education_level"])
```
